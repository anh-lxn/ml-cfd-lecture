# What is Basilisk? 

## Arguments for the Basilisk Solver

1. Maximum Refinement Level
  - the higher the number e.g. 14, the more cells (better details)
  - more cells = slower but more accurate
2. End Time (How long to simulate, e.g. 10 sec)
3. Eötvös number
  - relationship bouyancy (Auftrieb) and surface tension of the bubble
  - high Eötvös = bouyancy dominates -> bubble becomes more deformed
  - low Eötvös = surface tension dominates -> bubble stays more spherical (kugelförmig)
4. Galilei number
  - inertia (gravity driven) vs. viscosity
  - high Galilei = fast, turbelent movement of bubble
  - low Galilei = slow and smooth movement of bubble
