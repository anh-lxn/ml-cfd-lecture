```python
import matplotlib.pyplot as plt
import torch as pt
```

```python
results = {
  "test1" : pt.sin(pt.linspace(0, 7, 100)),
  "test2" : pt.cos(pt.linspace(0, 7, 100)),
}
```

```python
fig, ax  = plt.subplots(1, 2, sharey=True, sharex=True)
for i, (title, data) in enumerate(results.items()):
  ax[i].plot(pt.linspace(0, 7, 100), data)
  ax[i].set_title(title)
plt.show()
```
